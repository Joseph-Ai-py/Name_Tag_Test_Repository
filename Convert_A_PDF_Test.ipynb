```python
# ═══════════════════════════════
# 색상 유틸
# ═══════════════════════════════
def hex_to_rgb(h):
    h = h.lstrip("#")
    return tuple(int(h[i:i+2], 16) for i in (0, 2, 4))

def darken_hex(c, f=0.52):
    r,g,b = hex_to_rgb(c)
    hue,lit,sat = colorsys.rgb_to_hls(r/255,g/255,b/255)
    lit = max(0.0,lit*f)
    r2,g2,b2 = colorsys.hls_to_rgb(hue,lit,sat)
    return "#{:02X}{:02X}{:02X}".format(int(r2*255),int(g2*255),int(b2*255))

def lighten_hex(c, f=1.22):
    r,g,b = hex_to_rgb(c)
    hue,lit,sat = colorsys.rgb_to_hls(r/255,g/255,b/255)
    lit = min(1.0,lit*f)
    r2,g2,b2 = colorsys.hls_to_rgb(hue,lit,sat*0.35)
    return "#{:02X}{:02X}{:02X}".format(int(r2*255),int(g2*255),int(b2*255))
```

```python
# ═══════════════════════════════
# HTML 이스케이프
# ═══════════════════════════════
def e(text):
    """안전한 HTML 이스케이프 + 줄바꿈 처리."""
    return (str(text) if text else "") \
        .replace("&","&amp;").replace("<","&lt;") \
        .replace(">","&gt;").replace("\n","<br>")
```

```python
# ═══════════════════════════════
# 메인 조립 함수
# ═══════════════════════════════
def generate_html(data):
    """
    브랜드 dict → WeasyPrint용 완성된 HTML 반환.

    Args:
        data (dict): 브랜드 JSON 데이터

    Returns:
        str: 완성된 HTML 문자열
    """
    seed  = data.get("seed_color", "#F5EBE0")
    sd    = darken_hex(seed)
    sl    = lighten_hex(seed)
    brand = data.get("brand_name", "Brand Guide")

    css = build_css(seed, sd, sl)
    content = (
        build_cover(data)
        + build_basic_info(data)
        + build_a1_philosophy(data)
        + build_a2_essence(data)
        # + build_color_section(data, seed, sd, sl)
        + build_a3_mvv(data)
        + build_a4_naming(data)
        + build_a5_slogans(data)
        + build_a6_story(data)
        + build_a7_positioning(data)
        + build_a8_promise(data)
    )

    return f"""<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8"/>
  <title>{brand} 브랜드 아이덴티티 가이드</title>
  <style>{css}</style>
</head>
<body>
{content}
</body>
</html>"""
```

```python
# ═══════════════════════════════
# PDF 변환
# ═══════════════════════════════
def html_to_pdf(html_content, output_path):
    """weasyprint로 HTML → PDF 변환."""
    try:
        from weasyprint import HTML
        HTML(string=html_content).write_pdf(output_path)
        print(f"✅ PDF 저장 완료: {output_path}")
        return True
    except ImportError:
        print("❌ weasyprint 미설치  →  pip install weasyprint")
        return False
    except Exception as exc:
        print(f"❌ 변환 오류: {exc}")
        return False
```
